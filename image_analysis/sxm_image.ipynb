# Import modules

```python
from stm_utils import Sxm_Image
import numpy as np
import matplotlib.pyplot as plt
```

# Create file object

```python
file = "image.sxm"
im = Sxm_Image(file)
```

## Get attributes

```python
frame  = im.frame
print("frame size: ", frame)
```

Image metadata

```python
im.header
```

# Image

Default channel is Z_Fwd

```python
image = im.image()

plt.imshow(image, origin = 'lower')
plt.axis("off")

```

Get channels in the image

```python
channels = im.get_channels()
print(f"Channels present: {channels}")
```

Plot channels

```python
channel = "Z_Bkd"
ch_image = im.image(channel)

plt.imshow(image, origin = 'lower')
plt.axis("off")
```
